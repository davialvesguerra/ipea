```python
import pandas as pd
```

```python
data = pd.read_csv("dados_respondidos.csv")

topico = "tópico 6"

condicao_topico = data['topico'] == topico

condicao_dificuldade_2 = data['dificuldade'] == "2"
condicao_dificuldade_1 = (data['dificuldade'] == "1")*(data['resposta_user'] != data['resposta_certa'])
condicao_dificuldade = condicao_dificuldade_2 + condicao_dificuldade_1

id_topicos.append(data[(condicao_topico)*(condicao_dificuldade)].sample(n=10)["id"])
```

```python
# Assuming nested_list is your nested list
nested_list = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

# Flatten the list
flat_list = [item.to_list() for sublist in nested_list for item in id_topicos]
flat_list = list(set([item for sublist in flat_list for item in sublist]))

print(flat_list)  # Output: [1, 2, 3, 4, 5, 6, 7, 8, 9]
len(flat_list)
```

```python
data = pd.read_csv("dados_cru.csv")
data_a = data[data['id'].isin(flat_list)]
```

```python
data_b = data.iloc[-24:,:]
```

```python
data_c = pd.concat([data_a, data_b])
data_c.shape
```

```python
data_c.to_csv("dados_simulado.csv")
```
